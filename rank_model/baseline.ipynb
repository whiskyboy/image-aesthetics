```python
import cPickle as pickle

train_list = pickle.load(open("./data/train.list", 'rb'))
valid_list = pickle.load(open("./data/valid.list", 'rb'))
```

```python
imgCount = {}
for imgA, sA, imgB, sB, cmpret in train_list:
    imgCount.setdefault(imgA, [0, 0])
    imgCount.setdefault(imgB, [0, 0])
    imgCount[imgA][1] += 1
    imgCount[imgB][1] += 1
    if cmpret == 1:
        imgCount[imgA][0] += 1
    else:
        imgCount[imgB][0] += 1
```

```python
acc = 0
nohit = 0
hit = 0
Y_predict_class = []
for imgA, sA, imgB, sB, cmpret in valid_list:
    #assert(imgA not in imgCount)
    #assert(imgB in imgCount)
    if imgB not in imgCount:
        nohit += 1
        continue
    hit += 1
    p_B_win = imgCount[imgB][0] * 1.0 / imgCount[imgB][1]
    #print imgA, sA, imgB, sB, p_B_win
    if p_B_win > 0.5:
        cmp_predict = 0
        Y_predict_class.append(cmp_predict)
    else:
        cmp_predict = 1
        Y_predict_class.append(cmp_predict)
    if cmp_predict == cmpret:
        acc += 1

print acc * 1.0 / len(valid_list)
print nohit, hit
```

```python
p_score = {}
t_score = {}
for i, (img_a, s_a, img_b, s_b, _) in enumerate(valid_list):
    p_score.setdefault(img_a, 0)
    p_score[img_a] += Y_predict_class[i]
    t_score[img_a] = s_a
    
s1 = []
s2 = []
for img, s in sorted(p_score.items(), key=lambda x: x[1]):
    s1.append(s)
    s2.append(t_score[img])
    print img, s, t_score[img]
    
import numpy as np
from scipy.stats import spearmanr

print spearmanr(np.array(s1), np.array(s2))
```
